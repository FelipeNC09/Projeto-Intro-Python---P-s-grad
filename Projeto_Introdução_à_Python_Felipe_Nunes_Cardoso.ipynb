##**2. Códigos para aplicações usando dados *in vitro***  

####O sistema-modelo *in vitro* de glicação proteica baseia-se na incubação de albumina sérica bovina (BSA) com metilglioxal (MG), um composto dicarbonílico que age como agente agressor, por 4 dias, na ausência ou presença de diferentes concentrações de uma amostra de interesse e de aminoguanidina (AG), substância que atua como controle positivo de proteção.

####Ao longo dessa incubação, há o monitoramento espectrofluorimétrico da geração de diferentes níveis de marcadores do estresse glico-oxidativo, como produtos finais de glicação avançada (AGEs, do inglês "*advanced glycation end products*"), ditirosina, *N*'-formilquinurenina e quinurenina.

####A seguir, serão utilizados os dados referentes a uma incubação cuja amostra avaliada foi o extrato hidroetanólico de folhas de *Cissus verticillata* (CvExt).

###**2.1. Código para teste estatístico One-Way ANOVA**
